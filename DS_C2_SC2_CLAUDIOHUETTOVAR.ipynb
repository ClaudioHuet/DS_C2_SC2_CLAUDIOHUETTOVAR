```python
# MAYO'2023 / SCRIPT RETO PANDASYNUMPY / CLAUDIO HUET TOVAR

print("##### RESPUESTA INCISO A")
# a. Extraer la información del archivo en MyDrive
import pandas as pd
import numpy as np
mydf = pd.read_csv("/content/drive/MyDrive/country_vaccinations.csv")
# se pide propiedad shape como ejemplo de acceso al dataframe
print(mydf.shape)

print("##### RESPUESTA INCISO B")
# b. Mostrar la estructura y tipos de datos de cada columna para identificar 
# qué operaciones puedes realizar con cada una de ellas, asegurándote que las 
# columnas con fechas sean del tipo datetime64.

# para asegurar que las columnas con fechas sean del tipo datetime64
mydf["date"] = pd.to_datetime(mydf["date"])

# para mostrar nueva estructura con tipo datetime64 para columna date
mydf.info()

print("##### RESPUESTA INCISO C")
# c. Determinar la cantidad de vacunas aplicadas de cada compañía 
# (con base en cómo lo reporta cada país en la columna vaccines).
print(mydf[["vaccines", "daily_vaccinations"]].groupby(["vaccines"]).sum())


print("##### RESPUESTA INCISO D")
# d. Obtener la cantidad de vacunas aplicadas en todo el mundo. 
print(mydf["daily_vaccinations"].sum())

print("##### RESPUESTA INCISO E")
# e. Calcular el promedio de vacunas aplicadas por país.
# en este caso se calcula un promedio diario.
print(mydf[["country", "daily_vaccinations"]].groupby(["country"]).mean())

print("##### RESPUSTA INCISO F")
# f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.
# para convertir el texto a fecha para filtrarlo
mydate = pd.to_datetime("29/01/21")
# para filtrar el dataframe para la fecha requerida
myvacunas = mydf.loc[mydf["date"]==mydate,["date", "daily_vaccinations"]]
# para totalizar las vacunas aplicadas el dia 29/01/2021
print(myvacunas[["date", "daily_vaccinations"]].groupby(["date"]).sum())

print("##### RESPUESTA INCISO G EXPORTADA A EXCEL")
# g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos 
# originales y una columna derivada (diferencias) con las diferencias de 
# aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.

# para crear nuevo dataframe
conDiferencias = pd.read_csv("/content/drive/MyDrive/country_vaccinations.csv")
# para remplazar NA x ceros para poder sacar diferencias utilizando metodo nan de numpy y replace
conDiferencias["daily_vaccinations"] = conDiferencias["daily_vaccinations"].replace({np.nan:0})
conDiferencias["daily_vaccinations_raw"] = conDiferencias["daily_vaccinations_raw"].replace({np.nan:0})
# para crear columna derivada
conDiferencias["diferencias"] = conDiferencias["daily_vaccinations"]-conDiferencias["daily_vaccinations_raw"]


print("##### RESPUESTA INCISO H EXPRESADA EN MESES")
# h. Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.
# para obtener fecha mas reciente
mydatemax = mydf["date"].max()
# para obtener fecha más antigua
mydatemin = mydf["date"].min()
# para obtener periodo entre fechas
myperiodo = (mydatemax - mydatemin)
# para obtener el resultado en meses dividiendo con timedelta64 de numpy
myperiodo = myperiodo / np.timedelta64(1,"M")
print(f"{myperiodo} MESES")

print("##### RESPUESTA INCISO I EXPORTADA A EXCEL")
# i. Crear un dataframe nuevo denominado conCantidad que contenga los datos 
# originales y una columna derivada (canVac) con la cantidad de vacunas
# utilizadas cada día (usar la columna vaccines y separar por el carácter /). 

# para crear nuevo dataframe
conCantidad = pd.read_csv("/content/drive/MyDrive/country_vaccinations.csv")
# para separar columnas utilizadas utlizando separador /
myvaccines = conCantidad["vaccines"].str.split("/")
# para contar numero de vacunas
mycount = myvaccines.str.len()
# para crear columna derivada
conCantidad["canVac"] = mycount


print("##### RESPUESTA INCISO J EXPORTADA A EXCEL")
# j. Generar un dataframe denominado antes20 con todos los registros que 
# se hayan realizado antes del 20 de diciembre de 2020

# para crear nuevo dataframe
antes20 = mydf
# para convertir texto a date
mydate = pd.to_datetime("20/12/2020")
# para filtrar dataframe
antes20 = antes20[antes20["date"] < mydate]

print("##### RESPUESTA INCISO K EXPORTADA A EXCEL")
# k. Obtener un dataframe denominado pfizer con todos los registros donde 
# se haya utilizado la vacuna Pfizer.

# para crear nuevo dataframe
pfizer = mydf
mypfizer =pfizer["vaccines"].str.contains("Pfizer")
# para filtrar dataframe
pfizer = pfizer[(mypfizer==True)]

print("##### RESPUESTA INCISO L EXPORTADA A EXCEL")
# l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer)
# en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe
# ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter. 

#para crear mi excel
myexcel = pd.ExcelWriter("/content/drive/MyDrive/resultadosReto.xlsx")

#para guardar cada dataframe en pestaña diferente
conDiferencias.to_excel(myexcel, sheet_name="conDiferencias")
conCantidad.to_excel(myexcel, sheet_name="conCantidad")
antes20.to_excel(myexcel, sheet_name="antes20")
pfizer.to_excel(myexcel, sheet_name="pfizer")

# para cerrar el archivo con las pestañas correspondientes
myexcel.close()
```
